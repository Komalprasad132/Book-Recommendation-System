```python
df= books[['Title','Author','language_code','publication_date','publisher','genres']]
```

```python
df.head(1)
```

```python
df['Title']= df['Title'].str.split('(').str[0].str.strip()
```

```python
df.head()
```

```python
df['Author']= df['Author'].str.split('/')
```

```python
df.head()
```

```python
df['publication_date']= df['publication_date'].str.split('/').str[2]
```

```python
df['publication_date']= df['publication_date'].apply(lambda x: [x])
```

```python
df.head()
```

```python
others= df['language_code'].value_counts()[df['language_code'].value_counts() < 100]
df['language_code']= df['language_code'].apply(lambda x: ['other' if x in others else x])
```

```python
df.head()
```

```python
df['publisher']= df['publisher'].apply(lambda x: [x.replace(' ','')])
```

```python
# (df['genres'].apply(lambda x: set(x.split(';'))))
df['genres']= df['genres'].apply(lambda x:list(set(x.replace(" ", ";").replace(",",";").split(";"))))
```

```python
df.head()
```

```python
df['tags']= df['Author']+ df['language_code']+ df['publication_date']+ df['publisher']+ df['genres']
```

```python
df['tags'].iloc[0]
```

```python
new_df= df[['Title','tags']]
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['Title'][10]
```

```python
new_df1= new_df.iloc[0:1000] 
'''I have taken only 1k rows out of 11k because my system's memory was not able to fit all 11k rows in countVectorizer '''
```

```python
pd.options.display.max_rows= None
new_df1['Title']
```

```python
new_df1.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 5000)
```

```python
vectors= cv.fit_transform(new_df1['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
sort= sorted(list(enumerate(similarity[new_df1[new_df1['Title']=="The Hitchhiker's Guide to the Galaxy"].index[0]])), reverse= True, key= lambda x: x[1])[1:6]
for i in sort:
    book= new_df1['Title'].iloc[i[0]]
    print(book)
```

```python
new_df1[new_df1['Title']=="The Hitchhiker's Guide to the Galaxy"].index[0]
```

```python
def recommend(book):
    book_similarity= list(enumerate(similarity[new_df1[new_df1['Title']== book].index[0]])) 
    sorted_similarity= sorted(book_similarity, reverse= True, key= lambda x: x[1])[1:6]
    
    for i in sorted_similarity:
        book= new_df1['Title'].iloc[i[0]]
        print(book)
```

```python
recommend("The Lord of the Rings")
```
